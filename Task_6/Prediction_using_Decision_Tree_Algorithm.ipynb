```python
# Importing the required Libraries

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import train_test_split
import sklearn.metrics as sm

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import pydot
from IPython.display import Image
```

### Loading of Data

```python
# Loading Dataset
iris = load_iris()
X=iris.data[:,:]
y=iris.target
```

### Exploratory data analysis

```python
#Input data

data=pd.DataFrame(iris['data'],columns=["Petal length","Petal Width","Sepal Length","Sepal Width"])
data['Species']=iris['target']
data['Species']=data['Species'].apply(lambda x: iris['target_names'][x])

data.head()
```

```python
data.shape
```

```python
data.describe()
```

### Data Vizualization - Comparing of Various Columns

```python
# Input data Visualization
sns.pairplot(data)
```

```python
 # Scatter plot of data based on Sepal Length and Width features
sns.FacetGrid(data,hue='Species').map(plt.scatter,'Sepal Length','Sepal Width').add_legend()
plt.show()

# Scatter plot of data based on Petal Length and Width features
sns.FacetGrid(data,hue='Species').map(plt.scatter,'Petal length','Petal Width').add_legend()
plt.show()
```

### Decision Tree Model Training

```python
# Model Training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1)
tree_classifier = DecisionTreeClassifier()
tree_classifier.fit(X_train,y_train)
print("Training Complete.")
y_pred = tree_classifier.predict(X_test)
```

### Comparing Actual VS Prediction

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

### Vizualization - Trained Decision Tree Model

```python
#Visualizing the trained Decision Tree Classifier taking all 4 features in consideration

export_graphviz(
        tree_classifier,
        out_file="decision_tree.dot",
        feature_names=iris.feature_names[:],
        class_names=iris.target_names,
        rounded=True,
        filled=True
)

(graph,) = pydot.graph_from_dot_file('decision_tree.dot')
graph.write_png('decision_tree_image.png')

Image(filename='decision_tree_image.png')
```

###  Predicting the class output for some random values of petal and sepal length and width

```python
print("Class Names = ",iris.target_names)

# Estimating class probabilities
print()
print("Estimating Class Probabilities for flower whose petals length width are 4.7cm and 3.2cm and sepal length and width are 1.3cm and 0.2cm. ")
print()
print('Output = ',tree_classifier.predict([[4.7, 3.2, 1.3, 0.2]]))
print()
print("Our model predicts the class as 0, that is, setosa.")
```

### Calculating the Model Accuracy

```python
# Model Accuracy
print("Accuracy:",sm.accuracy_score(y_test, y_pred))
```

- This model has perfect accuracy (1 or 100%) because it uses all four features of the iris dataset.
- The decision tree model is based on the full set of iris features.
- All four iris features contribute to the 100% accuracy of this model.
